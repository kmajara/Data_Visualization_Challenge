### Analysis

This dataset consisted of around 250 mice assigned to different Cancer treatment regimens. The focus of the study was to compare performance of Capomulin agains the other treatment regimens. Capomulin has one of the highest subjects that participated in the study.  The median Tumor Volume observed for subjects on Capomulin was low, and only second to mice on Ramicane. Relatively low metrics were observed across key statistic as well, indicating the relatively efficacy of this regimen. 

The line plot visualizes the tumor volume dynamics over time for a single mouse (m601) under treatment with Capomulin. Notably, the tumor volume of mouse m601 exhibits a consistent decrease throughout the treatment period, indicative of a positive response to Capomulin. This observation aligns with the broader analysis of Capomulin's efficacy, as evidenced by the absence of potential outliers within the regimen's final tumor volume distribution. The plotted trend underscores Capomulin's effectiveness in mitigating tumor growth, portraying a promising outcome for the treatment regimen in this specific mouse model. A positive correlation was noted between the mice weight and average tumor volume. However, there is not sufficient information to conclude on causation. In conclusion, Capomulin is a promising regimen. Given the comparitive results for Ramicane, an additional evaluaion of the Ramicane should be performed before selecting the best regimen. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = study_results.merge(mouse_metadata, how = "left", on = ["Mouse ID"])

# Display the data table for preview
combined_data.head()
```

```python
# Check the number of mice.
combined_data["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

# Filter for duplicate rows considering both 'Mouse ID' and 'Timepoint'
duplicates = combined_data[combined_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep = 'first')]

# Extract Mouse IDs from duplicates
mouse_ids = duplicates['Mouse ID'].unique()
mouse_ids
```

```python
# Optional: showing data for the duplicated Mouse ID (only showing duplicate values in Mouse ID and Timepoint)
duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data[combined_data["Mouse ID"]!='g989'].copy()
clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_data["Mouse ID"].nunique()
```

## Summary Statistics

```python
# The following steps generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# First, use groupby to have "Drug Regimen" in the rows.
grouped_data = clean_data.groupby(["Drug Regimen"])

# Then, use summary statistical methods to calculate the following properties of each drug regimen:  mean, median, variance, standard deviation, and SEM of the tumor volume.
tumor_mean = grouped_data["Tumor Volume (mm3)"].mean()
tumor_median = grouped_data["Tumor Volume (mm3)"].median()
tumor_variance = grouped_data["Tumor Volume (mm3)"].var()
tumor_STD = grouped_data["Tumor Volume (mm3)"].std()
tumor_SEM = grouped_data["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.

summary_stats = pd.DataFrame({"Mean Tumor Volume": tumor_mean, "Median Tumor Volume": 
                              tumor_median, "Tumor Volume Variance": tumor_variance, 
                              "Tumor Volume Std. Dev": tumor_STD, "Tumor Volume Std. err": tumor_SEM})
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats_adv = grouped_data["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std","sem"])

#using a columns() function to rename columns for clarity

summary_stats_adv.columns = ["Mean Tumor Volume", "Median Tumor Volume", "Tumor Volume Variance", "Tumor Volume Std. Dev", "Tumor Volume Std. Err"]

#view stats
summary_stats_adv
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Put treatments into a list for for loop (and later for plot labels)
four_regimens_list = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

#filtering for desired regimens into a new dataframe
four_regimens_df = clean_data[clean_data['Drug Regimen'].isin(four_regimens_list)]

# Start by getting the last (greatest) timepoint for each mouse [created an additional column for Max Timepoint]

last_timepoint_df = pd.DataFrame(four_regimens_df.groupby('Mouse ID')['Timepoint'].max()).reset_index().rename(columns = {'Timepoint': 
                                                                                                                     'Max Timepoint'})
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_last_timepoint_df = pd.merge(four_regimens_df, last_timepoint_df, on = 'Mouse ID', how = "inner")

# view the result
merged_last_timepoint_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
four_regimens_list = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
 # Locate the rows which contain mice on each drug and get the tumor volumes
for drug in four_regimens_list:
    drug_df = merged_last_timepoint_df.loc[merged_last_timepoint_df["Drug Regimen"] == drug]

    # Extract tumor volume values
    max_tumor_df = drug_df.loc[drug_df["Timepoint"] == drug_df["Max Timepoint"], "Tumor Volume (mm3)"]
    tumor_vol_list.append(max_tumor_df)

    # Determine outliers using upper and lower bounds
    quartiles = max_tumor_df.quantile([.25, .75]).round(2)
    lower_quart = quartiles[0.25].round(2)
    upper_quart = quartiles[0.75].round(2)
    iqr = round(upper_quart - lower_quart, 2)
    lower_bound = round(lower_quart - (1.5 * iqr), 2)
    upper_bound = round(upper_quart + (1.5 * iqr), 2)

    # Identifying outliers
    outliers_df = max_tumor_df.loc[(max_tumor_df > upper_bound) | (max_tumor_df < lower_bound)]

    print("For " + str(drug) + ":, Upper bound:" + str(upper_bound))
    print("Lower bound:" + str(lower_bound))
    if len(outliers_df) != 0:
        print("Potential outliers for " + str(drug) + " are " + str(outliers_df))
    else:
        print("There are no potential outliers for " + str(drug))
    print("\n")
    
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
fig1, box = plt.subplots()

box.set_title("Tumor Volume Distribution")
plt.ylabel("Final Tumor Volume")
plt.xlabel("Drug Regimen")

box.boxplot(tumor_vol_list, 0, 'r')
box.boxplot(tumor_vol_list)
plt.xticks([1,2,3,4],four_regimens_list)
plt.grid()
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

#filter for capomulin in dataframe 
cap_df = four_regimens_df.loc[four_regimens_df["Drug Regimen"] == "Capomulin"]

# Plot line for m601 mouse
cap_m601_df = four_regimens_df.loc[four_regimens_df["Mouse ID"] == "m601"]
line_m601 = cap_m601_df.plot.line(x='Timepoint', y='Tumor Volume (mm3)', marker = "p")
plt.xlabel("Time (days)")
plt.grid()
plt.ylabel("Tumor Size (mm3)")
plt.title("Results of Mouse m601 on Capumolin")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Use previously created df for this analysis
# Group by 'Weight (g)' and calculate mean 'Tumor Volume (mm3)' for each group
import matplotlib.pyplot as plt

cap_group = cap_df.groupby('Mouse ID')
#calculated an average weight for the mice and avg tumor
weight = cap_group['Weight (g)'].mean()
avg_tumor = cap_group['Tumor Volume (mm3)'].mean()

# Create labels and titles
plt.scatter(weight, avg_tumor)
plt.xlabel("Weight (g)")
plt.grid()
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight vs. Average Tumor Volume for Mice on Capomulin")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

#cap_group contains the required columns to calculate the averages
cap_group = cap_df.groupby('Mouse ID')
weight = cap_group['Weight (g)'].mean()
avg_tumor = cap_group['Tumor Volume (mm3)'].mean()

plt.scatter(weight, avg_tumor)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight vs. Average Tumor Volume for Mice on Capomulin")

# Calculate the linear regression line
(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, avg_tumor)
reg = weight * slope + intercept

# Plot the regression line
plt.plot(weight, reg, color='red')

# Annotate the line equation
line_equation = f"y = {round(slope, 2)}x + {round(intercept, 2)}"
plt.annotate(line_equation, (18, 38), fontsize=12, color="red")

plt.grid()
plt.show()

# Calculate and print the correlation coefficient
cap_cor = np.corrcoef(weight, avg_tumor)[0, 1]
print("The correlation between mouse weight and average tumor volume is", round(cap_cor, 2))
print(line_equation)
```
